```python
# Arithmatic operations 


x = float(input("value of x: "))
y = float(input("value of y: "))

sum = x + y
product = x*y 
difference = x - y 
quotient = x/y
remainder = x%y

arithematic_results = [("sum",sum), ("product",product), ("difference",difference) , ("quotient",quotient) , ("remainder",remainder)]

for label,result in arithematic_results:
    print(f"{label}: {result}")
```

```python
def even_or_odd():
    while True:
        num = input("Enter a positive integer to check or the word 'End' to end ==>  ").strip()

        if num.lower() == 'end':
            print("Bye bye")
            break

        elif num.isdigit():
            num = int(num)
            if num > 0:  #  ensures positive integer
                print(f"The {num} is {'even' if num % 2 == 0 else 'odd'}")
            else:
                print("Zero is not a positive integer!")
        else:
            print(f"Do you really think {num} is a positive integer?")

# call the function
even_or_odd()
```

```python
def hogwarts_records():
    alumni_data = {
        "Gryffindor": ("Ron", "Hermione", "Harry"),
        "Slytherin": ("Draco", "Snape"),
        "Ravenclaw": ("Luna", "Cho"),
        "Hufflepuff": ("Cedric", "Ginny")
    }

    name = input("Enter the student Name: ").strip().title()

    print(
        next(
            (f"{name} is from the house: {house}" 
             for house, students in alumni_data.items() if name in students),
            f"{name} is not in Hogwarts records."
        )
    )

# Call the function
hogwarts_records()
```
