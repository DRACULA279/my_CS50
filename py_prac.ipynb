```python
def even_or_odd():
    while True:
        num = input("Enter a positive integer to check or the word 'End' to end ==>  ").strip()

        if num.lower() == 'end':
            print("Bye bye")
            break

        elif num.isdigit():
            num = int(num)
            if num > 0:  #  ensures positive integer
                print(f"The {num} is {'even' if num % 2 == 0 else 'odd'}")
            else:
                print("Zero is not a positive integer!")
        else:
            print(f"Do you really think {num} is a positive integer?")

# call the function
even_or_odd()
```

```python
import random

# Rock, Paper, Scissors Game

choices = ["rock", "paper", "scissors"]

user_score = 0
computer_score = 0

while True:
    user_choice = input("Enter rock, paper, scissors or 'stop' to end: ").strip().lower()
    computer_choice = random.choice(choices)

    if user_choice == "stop":
        print(f"\nGame Ended. Thank You.\nUser Score: {user_score}\nComputer Score: {computer_score}")
        break

    if user_choice not in choices:
        print(f"Invalid choice! Please choose from {choices}.")
        continue

    print(f"Computer chose: {computer_choice}")

    if (
        (user_choice == "rock" and computer_choice == "scissors") or
       (user_choice == "paper" and computer_choice == "rock") or 
       (user_choice == "scissors" and computer_choice == "paper")
       
    ):
        user_score += 1
        print(f"You won this round!\nUser Score: {user_score}\nComputer Score: {computer_score}")
    elif user_choice == computer_choice:
        print(f"It's a tie!\nUser Score: {user_score}\nComputer Score: {computer_score}")
    else:
        computer_score += 1
        print(f"Computer won this round!\nUser Score: {user_score}\nComputer Score: {computer_score}")
```

```python
#A triangle can be classified based on the lengths of its sides as equilateral, isosceles or scalene. All 3 sides of an equilateral triangle have the same length. An isosceles triangle has two sides that are the same length, and a third side that is a different length. If all of the sides have different lengths then the triangle is scalene. Write a program that reads the lengths of 3 sides of a triangle from the user. Display a message indicating the type of the triangle.

def type_of_triangle():

    adjacent = float(input("What's the length of adjacent side?: "))
    hypotenuse = float(input("What's the length of hypotenuse side?: "))
    opposite = float(input("What's the length of opposite side?: "))

    if adjacent == opposite == hypotenuse:
            print("The triangle is Equilateral")

    elif adjacent != opposite  and opposite != hypotenuse and adjacent != hypotenuse:
            print(" The Triangle is Scalene.")

    else:
         print("The triangle is Isosceles.")  
    
type_of_triangle()
```

```python
def prime_checker():
    num = int(input("Enter a number to check if it's a prime: "))

    if num <= 1 :
        print("Enter a positive integer")
        return
    
    for i in range(2, num):
        if num % i == 0:
            print(f"{num} is not prime")
            return

    print(f"{num} is prime")

prime_checker()
```

```python
#
def gnome_sort(arr):
    i = 0
    n = len(arr)
    while i < n:
        # if second number >= first number the list stays the same and the loop continues
        if i == 0 or arr[i] >= arr[i-1]:
            i += 1

        # if not the numbers will get interchanged         
        else:
            arr[i], arr[i-1] = arr[i-1],arr[i]
            i -= 1

    return arr
data = [2,4,56,176,54,6,2]
gnome_sort(data)
```

```python
def vowels():
    letter = input("Enter a letter: ").strip().lower()

    if len(letter) != 1 or not letter.isalpha():
        print("Please enter a single alphabetic character.")
        return

    match letter:
        case 'a' | 'e' | 'i' | 'o' | 'u':
            print(f"The letter {letter} is a vowel.")
        case _:
            print(f"The letter {letter} is a consonant.")

vowels()
```

```python
# . The length of a month varies from 28 to 31 days. Create a program that reads the name of a month from the user as a string. Then your program should display the number of days in that month. Display “28 or 29 days” for February so that leap years is addressed.

def calendar():

    month = input("Enter the month you need to know the number of days it got: ").lower().strip()

    match month:
        case "january" | "march" |"may" | "july" |"august" | "october" |"december":
            print(f"The month {month} has 31 days")
        
        case 'february':
            print("The month {month} has 29 days in a leap year and 28 for normal year.")

        case "april" | 'june' | 'september' | 'november':
            print(f"The month {month} has 30 days.")

        case _:
            print(f"{month} is not a month, clearly.")
              

calendar()
```

```python
from random import choice

def game():
    choices = ['rock', 'paper', 'scissors']
    user_score = 0
    computer_score = 0

    while user_score < 5 and computer_score < 5:

        user_choice = input(f"Choose from {choices}: ").lower().strip()

        if user_choice not in choices:
            print(f"Invalid input. Choose from {choices}")
            continue

        computer_choice = choice(choices)

        print(f"Computer chose {computer_choice}, you chose {user_choice}")

        if user_choice == computer_choice:
            print("It's a tie.")

        elif (
            (user_choice == 'rock' and computer_choice == 'scissors') or
            (user_choice == 'paper' and computer_choice == 'rock') or
            (user_choice == 'scissors' and computer_choice == 'paper')
        ):
            user_score += 1
            print("You win this round!")

        else:
            computer_score += 1
            print("Computer won this round!")

        print(f"Score :\nYou: {user_score} | Computer: {computer_score}\n")

    # Final result
    if user_score == 5:
        print("Congrats, you won the tournament.")
    else:
        print("You lost. Better luck next time.")

game()
```

```python
#"Write a program to print to find the sum of series .1 + 2 + 6 + 24 +120 … n terms."


from math import factorial

def factorial_sum():

    n = int(input("Enter number of terms: "))

    
    for i in range(1, n+1):
        
        sum = factorial(i)
    
    print(sum)

factorial_sum()
```
